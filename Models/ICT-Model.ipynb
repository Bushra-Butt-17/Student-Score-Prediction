```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# Features (Scores of activities in the morning data)
features = ['Q1',	'Q2',	'A1',	'Q3',	'Q4',	'Midterm',	'Q5',	'A2',	'Q6',	'Q7',	'Q8',	'Final']
# Select features and label (Final score) from the morning data
X_train = morning_data[features]  # Features for training
y_train = morning_data['Total']  # Final score for training

# Select the features from the afternoon data (testing set)
X_test = afternoon_data[features]  # Features for testing
y_test = afternoon_data['Total']  # Actual final score for testing

# Initialize and train a Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the afternoon data
y_pred = model.predict(X_test)

# Evaluate the model
mae = mean_absolute_error(y_test, y_pred)  # Mean Absolute Error
mse = mean_squared_error(y_test, y_pred)  # Mean Squared Error
r2 = r2_score(y_test, y_pred)  # R-squared (goodness of fit)

# Output the evaluation results
print(f"Mean Absolute Error: {mae}")
print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")

# Optionally, print a comparison of predicted vs actual scores
comparison_df = pd.DataFrame({'Predicted': y_pred, 'Actual': y_test})
print(comparison_df.head())
```

```python
# Define the list of all activities (column names related to activities)
all_activities = ['Q1', 'Q2', 'A1', 'Q3', 'Q4', 'Midterm', 'Q5', 'A2', 'Q6', 'Q7', 'Q8', 'Final']

# Calculate the total number of activities
activity_count = len(all_activities)  # Total number of activity columns

# Define the function to build features dynamically up to the nth activity
def build_features_for_prediction(data, num_activities):
    # Select features dynamically based on the number of activities
    features = all_activities[:num_activities]
    return features  # Return the feature names as a list

# Train the model for each step after the 5th activity
predictions = []
for num_activities in range(5, activity_count + 1):
    # Select features for training based on the number of activities
    X_train = morning_data[build_features_for_prediction(morning_data, num_activities)]
    y_train = morning_data['Total']
    
    # Select features for testing based on the number of activities
    X_test = afternoon_data[build_features_for_prediction(afternoon_data, num_activities)]
    y_test = afternoon_data['Total']
    
    # Initialize and train the Linear Regression model
    model = LinearRegression()
    model.fit(X_train, y_train)
    
    # Make predictions on the testing set
    y_pred = model.predict(X_test)
    
    # Evaluate the model performance
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    # Store the predictions and evaluation results
    predictions.append({
        'num_activities': num_activities,
        'mae': mae,
        'mse': mse,
        'r2': r2,
        'predictions': y_pred
    })
    
    # Print the evaluation for each step
    print(f"Evaluation for prediction after {num_activities} activities:")
    print(f"Mean Absolute Error: {mae}")
    print(f"Mean Squared Error: {mse}")
    print(f"R-squared: {r2}")
    
    # Optionally, print a comparison of predicted vs actual scores
    comparison_df = pd.DataFrame({'Predicted': y_pred, 'Actual': y_test})
    print(comparison_df.head())

# Optionally, visualize the predictions for each activity step
import matplotlib.pyplot as plt

for result in predictions:
    plt.plot(result['predictions'], label=f"After {result['num_activities']} activities")
    
plt.xlabel('Student Index')
plt.ylabel('Predicted Total Score')
plt.title('Predictions for Each Activity Step')
plt.legend()
plt.show()
```

```python
import numpy as np
# Train the model for each step after the 5th activity
predictions = []
for num_activities in range(5, activity_count + 1):
    # Select features for training based on the number of activities
    X_train = morning_data[build_features_for_prediction(morning_data, num_activities)]
    y_train = morning_data['Total']
    
    # Select features for testing based on the number of activities
    X_test = afternoon_data[build_features_for_prediction(afternoon_data, num_activities)]
    y_test = afternoon_data['Total']
    
    # Initialize and train the Linear Regression model
    model = LinearRegression()
    model.fit(X_train, y_train)
    
    # Make predictions on the testing set
    y_pred = model.predict(X_test)
    
    # Round the predictions to the nearest integer
    y_pred_rounded = np.round(y_pred)
    
    # Evaluate the model performance (without rounding)
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    # Evaluate the model performance (with rounding)
    mae_rounded = mean_absolute_error(y_test, y_pred_rounded)
    mse_rounded = mean_squared_error(y_test, y_pred_rounded)
    r2_rounded = r2_score(y_test, y_pred_rounded)
    
    # Store the predictions and evaluation results
    predictions.append({
        'num_activities': num_activities,
        'mae': mae,
        'mse': mse,
        'r2': r2,
        'mae_rounded': mae_rounded,
        'mse_rounded': mse_rounded,
        'r2_rounded': r2_rounded,
        'predictions': y_pred,
        'predictions_rounded': y_pred_rounded
    })
    
    # Print the evaluation for each step
    print(f"Evaluation for prediction after {num_activities} activities (Without Rounding):")
    print(f"Mean Absolute Error: {mae}")
    print(f"Mean Squared Error: {mse}")
    print(f"R-squared: {r2}")
    print(f"Evaluation for prediction after {num_activities} activities (With Rounding):")
    print(f"Mean Absolute Error: {mae_rounded}")
    print(f"Mean Squared Error: {mse_rounded}")
    print(f"R-squared: {r2_rounded}")
    
    # Optionally, print a comparison of predicted vs actual scores
    comparison_df = pd.DataFrame({'Predicted': y_pred, 'Rounded': y_pred_rounded, 'Actual': y_test})
    print(comparison_df.head())

# Optionally, visualize the predictions for each activity step (with and without rounding)
import matplotlib.pyplot as plt

for result in predictions:
    plt.plot(result['predictions'], label=f"After {result['num_activities']} activities (Original)")
    plt.plot(result['predictions_rounded'], label=f"After {result['num_activities']} activities (Rounded)")

plt.xlabel('Student Index')
plt.ylabel('Predicted Total Score')
plt.title('Predictions for Each Activity Step (With and Without Rounding)')
plt.legend()
plt.show()
```

```python
# Define the list of all activities (column names related to activities)
all_activities = ['Q1', 'Q2', 'A1', 'Q3', 'Q4', 'Midterm', 'Q5', 'A2', 'Q6', 'Q7', 'Q8', 'Final']

# Calculate the total number of activities
activity_count = len(all_activities)  # Total number of activity columns

# Define the function to build features dynamically up to the nth activity
def build_features_for_prediction(data, num_activities):
    # Select features dynamically based on the number of activities
    features = all_activities[:num_activities]
    return features  # Return the feature names as a list

# Train the model for all steps after the 5th activity
predictions = {}
for num_activities in range(5, activity_count + 1):
    # Select features for training and testing based on the number of activities
    X_train = morning_data[build_features_for_prediction(morning_data, num_activities)]
    y_train = morning_data['Total']
    X_test = afternoon_data[build_features_for_prediction(afternoon_data, num_activities)]
    y_test = afternoon_data['Total']

    # Initialize and train the Linear Regression model
    model = LinearRegression()
    model.fit(X_train, y_train)

    # Make predictions on the testing set
    y_pred = model.predict(X_test)

    # Store the predictions for the given number of activities
    predictions[num_activities] = {
        'predicted': y_pred,
        'actual': y_test.to_numpy()
    }

# Create an interactive function for user input
def predict_final_score():
    while True:
        try:
            num_activities = int(input(f"Enter the number of activities (5-{activity_count}) to predict the scores: "))
            if num_activities < 5 or num_activities > activity_count:
                print(f"Please enter a value between 5 and {activity_count}.")
                continue
            
            # Retrieve the predictions for the selected number of activities
            result = predictions[num_activities]
            predicted_scores = result['predicted']
            actual_scores = result['actual']

            # Display the predictions
            print(f"\nPredicted Scores After {num_activities} Activities:")
            print(predicted_scores)

            print(f"\nActual Scores After {num_activities} Activities:")
            print(actual_scores)

            # Visualize the predicted vs actual scores
            import matplotlib.pyplot as plt

            plt.figure(figsize=(10, 6))
            plt.plot(actual_scores, label='Actual', marker='o', color='blue')
            plt.plot(predicted_scores, label='Predicted', marker='x', color='red')
            plt.title(f"Actual vs Predicted Scores (After {num_activities} Activities)")
            plt.xlabel("Student Index")
            plt.ylabel("Total Score")
            plt.legend()
            plt.grid(True)
            plt.show()

            # Ask if the user wants to continue
            cont = input("Would you like to check predictions for another number of activities? (yes/no): ").lower()
            if cont != 'yes':
                break

        except ValueError:
            print("Invalid input. Please enter a valid number.")

# Call the interactive function
predict_final_score()
```
