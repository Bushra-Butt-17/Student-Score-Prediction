```python
import pandas as pd

# Load the Excel file
file_path = 'CC.xlsx'  # Replace with the actual file path of your Excel file
excel_data = pd.ExcelFile(file_path)

# Check sheet names to confirm the structure
print(excel_data.sheet_names)

# Load data from the Morning and Afternoon sheets
morning_data = excel_data.parse('Morning')
afternoon_data = excel_data.parse('Afternoon')

# Display the first few rows of both datasets to inspect them
print("Morning Data Sample:")
print(morning_data.head())

print("\nAfternoon Data Sample:")
print(afternoon_data.head())
```

```python
# Drop irrelevant columns (e.g., Weights/Scale, Unnamed columns)
columns_to_drop = ['Weights/Scale', 'Unnamed: 1', 'Unnamed: 2']
morning_data.drop(columns=columns_to_drop, inplace=True)
afternoon_data.drop(columns=columns_to_drop, inplace=True)

# Fill missing values (NaNs) with a strategy of your choice, e.g., 0 or mean (here we use 0 for simplicity)
morning_data.fillna(0, inplace=True)
afternoon_data.fillna(0, inplace=True)

# Display cleaned morning and afternoon data samples
print("Cleaned Morning Data Sample:")
print(morning_data.head())

print("\nCleaned Afternoon Data Sample:")
print(afternoon_data.head())
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# Features (Scores of activities in the morning data)
features = ['A1', 'Q1', 'A2', 'Q2', 'A3', 'A4', 'Q3', 'Mid', 'AWS Labs', 'Q4', 'A5', 'Q5', 'A6']

# Select features and label (Final score) from the morning data
X_train = morning_data[features]  # Features for training
y_train = morning_data['Total']  # Final score for training

# Select the features from the afternoon data (testing set)
X_test = afternoon_data[features]  # Features for testing
y_test = afternoon_data['Total']  # Actual final score for testing

# Initialize and train a Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the afternoon data
y_pred = model.predict(X_test)

# Evaluate the model
mae = mean_absolute_error(y_test, y_pred)  # Mean Absolute Error
mse = mean_squared_error(y_test, y_pred)  # Mean Squared Error
r2 = r2_score(y_test, y_pred)  # R-squared (goodness of fit)

# Output the evaluation results
print(f"Mean Absolute Error: {mae}")
print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")

# Optionally, print a comparison of predicted vs actual scores
comparison_df = pd.DataFrame({'Predicted': y_pred, 'Actual': y_test})
print(comparison_df.head())
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt
import pandas as pd

# Assume the data is already loaded in `morning_data` and `afternoon_data`

# Define all activities column names (correct based on your dataset)
all_activities = ['A1', 'Q1', 'A2', 'Q2', 'A3', 'A4', 'Q3', 'Mid', 'AWS Labs', 'Q4', 'A5', 'Q5', 'A6','Final']

# Calculate the total number of activities
activity_count = len(features)

# Define the function to build features dynamically up to the nth activity
def build_features_for_prediction(data, num_activities):
    features = all_activities[:num_activities]  # Select features dynamically
    return features

# Train the model for each step after the 5th activity
predictions = []
for num_activities in range(5, activity_count + 1):
    try:
        # Select features for training based on the number of activities
        X_train = morning_data[build_features_for_prediction(morning_data, num_activities)]
        y_train = morning_data['Total']

        # Select features for testing based on the number of activities
        X_test = afternoon_data[build_features_for_prediction(afternoon_data, num_activities)]
        y_test = afternoon_data['Total']

        # Initialize and train the Linear Regression model
        model = LinearRegression()
        model.fit(X_train, y_train)

        # Make predictions on the testing set
        y_pred = model.predict(X_test)

        # Evaluate the model performance
        mae = mean_absolute_error(y_test, y_pred)
        mse = mean_squared_error(y_test, y_pred)
        r2 = r2_score(y_test, y_pred)

        # Store the predictions and evaluation results
        predictions.append({
            'num_activities': num_activities,
            'mae': mae,
            'mse': mse,
            'r2': r2,
            'predictions': y_pred
        })

        # Print the evaluation for each step
        print(f"Evaluation for prediction after {num_activities} activities:")
        print(f"Mean Absolute Error: {mae}")
        print(f"Mean Squared Error: {mse}")
        print(f"R-squared: {r2}")

        # Optionally, print a comparison of predicted vs actual scores
        comparison_df = pd.DataFrame({'Predicted': y_pred, 'Actual': y_test})
        print(comparison_df.head())
        
    except KeyError as e:
        print(f"Error: Missing columns for {num_activities} activities: {e}")

# Optionally, visualize the predictions for each activity step
plt.figure(figsize=(10, 6))
for result in predictions:
    plt.plot(result['predictions'], label=f"After {result['num_activities']} activities")

plt.xlabel('Student Index')
plt.ylabel('Predicted Total Score')
plt.title('Predictions for Each Activity Step')
plt.legend()
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

# Extract data for visualization
num_activities = [result['num_activities'] for result in predictions]
mae_values = [result['mae'] for result in predictions]
mse_values = [result['mse'] for result in predictions]
r2_values = [result['r2'] for result in predictions]

# Create a figure for visualizations
plt.figure(figsize=(15, 10))

# Plot MAE over number of activities
plt.subplot(2, 2, 1)
plt.plot(num_activities, mae_values, marker='o', label="Mean Absolute Error (MAE)", color='blue')
plt.title("MAE vs. Number of Activities")
plt.xlabel("Number of Activities")
plt.ylabel("Mean Absolute Error")
plt.grid(True)
plt.legend()

# Plot MSE over number of activities
plt.subplot(2, 2, 2)
plt.plot(num_activities, mse_values, marker='o', label="Mean Squared Error (MSE)", color='orange')
plt.title("MSE vs. Number of Activities")
plt.xlabel("Number of Activities")
plt.ylabel("Mean Squared Error")
plt.grid(True)
plt.legend()

# Plot R-squared over number of activities
plt.subplot(2, 2, 3)
plt.plot(num_activities, r2_values, marker='o', label="R-squared (R2)", color='green')
plt.title("R-squared vs. Number of Activities")
plt.xlabel("Number of Activities")
plt.ylabel("R-squared")
plt.grid(True)
plt.legend()

# Compare Predicted vs Actual for a selected activity count (e.g., 8 activities)
selected_activity = 8
selected_prediction = next(result for result in predictions if result['num_activities'] == selected_activity)
predicted = selected_prediction['predictions']
actual = afternoon_data['Total'].to_numpy()

plt.subplot(2, 2, 4)
plt.plot(actual, label='Actual', marker='o', color='blue')
plt.plot(predicted, label=f'Predicted (After {selected_activity} activities)', marker='x', color='red')
plt.title(f"Actual vs Predicted Scores (After {selected_activity} Activities)")
plt.xlabel("Student Index")
plt.ylabel("Total Score")
plt.grid(True)
plt.legend()

# Adjust layout and show plots
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Create a summary table to visualize the differences
summary_data = []

for result in predictions:
    num_activities = result['num_activities']
    y_pred = result['predictions']
    y_actual = afternoon_data['Total'].to_numpy()
    
    # Calculate the difference (error) for each student
    difference = y_actual - y_pred
    summary_data.append({
        'Num Activities': num_activities,
        'MAE': result['mae'],
        'MSE': result['mse'],
        'R2': result['r2'],
        'Avg Diff': difference.mean(),  # Average difference
        'Max Diff': difference.max(),  # Maximum difference
        'Min Diff': difference.min()   # Minimum difference
    })

# Convert to a DataFrame for easy visualization
summary_df = pd.DataFrame(summary_data)

# Display the summary table
print(summary_df)

# Optionally, save to a CSV file
summary_df.to_csv('prediction_summary.csv', index=False)
```
